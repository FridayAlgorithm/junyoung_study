### [우선순위 큐] [최대 힙](https://www.acmicpc.net/problem/11279)
널리 잘 알려진 자료구조 중 최대 힙이 있다. 최대 힙을 이용하여 다음과 같은 연산을 지원하는 프로그램을 작성하시오.

1. 배열에 자연수 x를 넣는다.
2. 배열에서 가장 큰 값을 출력하고, 그 값을 배열에서 제거한다.

프로그램은 처음에 비어있는 배열에서 시작하게 된다.

```python
import sys

n = int(sys.stdin.readline())
heap = []
heap.append(0)

'''
 최대 힙 삽입 연산 :
 삽입하려는 값을 맨 마지막 노드로 넣어주고 부모 노드의 값과 비교하여
 자식 노드가 부모 노드보다 크지 않을 때까지 자리를 바꾼다.
 
 자식 노드 : x
 부모 노드 : x//2
'''
def insert(x):
    heap.append(x)
    i = len(heap)-1
    
    while i>1:
        if heap[i//2]<heap[i]:
            tmp = heap[i]
            heap[i] = heap[i//2]
            heap[i//2] = tmp
            i = i//2
        else:
            break

'''
 최대 힙 삭제 연산 :
 맨 마지막 노드를 제일 큰 값이 있는 1번 노드에 삽입하고 자식 노드들의 값과 비교하여 
 부모 노드가 자식 노드보다 작지 않을 때까지 자리를 바꾼다.
 이때, 왼쪽 자식 노드와 오른쪽 자식 노드의 값을 비교하여 더 큰 자식 노드와 바꾼다.
 
 부모 노드 : x
 자식 노드 : x*2(왼쪽), x*2+1(오른쪽)
'''
def remove():
    if len(heap) == 1:
        print(0)
    else:
        print(heap[1])

        top = heap.pop()
        parent = 1
        child = 2
        
        if len(heap) != 1:
            heap[parent] = top

        while child <= len(heap)-1:
            if child < len(heap)-1 and heap[child] < heap[child+1]:
                child += 1
                
            if heap[child] <= top:
                break
                
            tmp = heap[parent]
            heap[parent] = heap[child]
            heap[child] = tmp

            parent = child
            child *= 2
            
while n>0:
    n-=1
    order = int(sys.stdin.readline())
    if(order == 0):
        remove()
    else:
        insert(order)
        
```

```python
import heapq
'''
 heapq는 최소힙으로 구현되어 있다.
 따라서 입력 받은 값에 음수 값을 취해 push, pop을 해야 한다.
'''

n = int(input())
heap = []

for _ in range(n):
    num = int(input())
    if num == 0:
        try:
            print(-1 * heapq.heappop(heap))
        except:
            print(0)
    else:
        heapq.heappush(heap, -num)
```
