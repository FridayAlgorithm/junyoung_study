```python
from itertools import permutations

def solution(expression):
    answer = 0
    
    operations = list(permutations('*+-', 3))
    for op in operations:
        if expression.isdigit():
            result = abs(expression)
        else:
            exp = list(expression.split(op[2]))
            print(exp)
            for i in range(len(exp)):
                exp[i] = exp[i].split(op[1])
            
            for i in range(len(exp)):
                for j in range(len(exp[i])):
                    if op[0] in exp[i][j]:
                        exp[i][j] = "("+exp[i][j]+")"
                exp[i][0] = "("+exp[i][0]
                exp[i][-1] = exp[i][-1]+")"
            print(exp)
                
            for i in range(len(exp)):
                exp[i] = op[1].join(exp[i])
            print(exp)
            
            exp = op[2].join(exp)
            print(exp)
            answer = max(answer, abs(eval(exp)))
            
    return answer

print(solution("100-200*300-500+20"))
print(solution("50*6-3*2"))
```
