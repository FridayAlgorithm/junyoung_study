### [재귀] [별 찍기 - 10](https://www.acmicpc.net/problem/2447)

재귀적인 패턴으로 별을 찍어 보자. N이 3의 거듭제곱(3, 9, 27, ...)이라고 할 때, 크기 N의 패턴은 N×N 정사각형 모양이다.

크기 3의 패턴은 가운데에 공백이 있고, 가운데를 제외한 모든 칸에 별이 하나씩 있는 패턴이다.
N이 3보다 클 경우, 크기 N의 패턴은 공백으로 채워진 가운데의 (N/3)×(N/3) 정사각형을 크기 N/3의 패턴으로 둘러싼 형태이다. 예를 들어 크기 27의 패턴은 예제 출력 1과 같다.

```python
n = int(input())

def print_star(n):
    for i in range(1, n+1):
        for j in range(1, n+1):
            if i > n/3 and i<= n*2/3 and j> n/3 and j<= n*2/3:
                print(" ", end="")
            else:
                print("*", end="")
            print()

print_star(n)
```

```python
n = int(input())
stars = [[" "]*n for _ in range(n)]

def print_stars(size, x, y):
    if size == 1:
        stars[y][x] = "*"
    else:
        next_size = size//3
        for next_y in range(3):
            for next_x in range(3):
                if not(next_y == 1 and next_x == 1):
                    print_stars(next_size, x+next_x*next_size, y+next_y*next_size)
                    
print_stars(n, 0, 0)
for i in stars:
    print(''.join(i))
```
