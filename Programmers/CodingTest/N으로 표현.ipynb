## N으로 표현
https://programmers.co.kr/learn/courses/30/lessons/42895?language=python3

* 동적계획법 : 인덱스가 N을 몇 번 사용했는지를 나타냄

[이해]
* N(5) - 1번 : 5
* N(5) - 2번 : 55, 5+5, 5-5, 5*5, 5//5
* N(5) - 3번 : 555, 5+55, 5-55, ... 

```python
def solution(N, number):
    dp = [[]]
    for i in range(1, 9):
        case = [int(str(N)*i)]
        for j in range(1, i):
            for x in dp[j]:
                for y in dp[i-j]:
                    case.append(x+y)
                    case.append(x-y)
                    case.append(x*y)
                    if y != 0:
                        case.append(x//y)
        print(case)
        if number in case:
            return i
        dp.append(case)
            
    return -1

print(solution(5, 12))
print(solution(2, 11))
```
